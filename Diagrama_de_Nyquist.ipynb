\begin{align}G(s) = \frac{1}{s-0.5}
\end{align}

```python
from pylab import plot
import control
from control import TransferFunction

s = TransferFunction.s
G  = (2)/(2*s - 1)
control.nyquist(G)

plot()
```

\begin{align}G(s) = \frac{1}{(s+1)(s-0.5)}
\end{align}

```python
G  = (1)/((s+1)*(s-0.5))
control.nyquist(G)

plot()
```

\begin{align}G(s) = \frac{s+1}{s-0.5}
\end{align}

```python
G  = (s+1)/(s-0.5)
control.nyquist(G)

plot()
```

\begin{align}G(s) = \frac{1}{s^{2}(s+1)(s+0.5)}
\end{align}

```python
G  = (1)/(s**2)*(s+1)*(s+0.5)
control.nyquist(G)

plot()
```

\begin{align}G(s) = \frac{s}{(s+1)(s+0.5)}
\end{align}

```python
G  = (s)/(s+1)*(s+0.5)
control.nyquist(G)

plot()
```

# Funções utilizadas:

```python
import numpy as np


def F(x):
  y = np.zeros(2)
  y[0] = (np.cos(x[0])/(x[0]+1) - np.sin(x[0])/(1+x[0])**2 - np.cos(x[1])/(1+x[1]) + np.sin(x[1])/(1+x[1])**2)
  y[1] = (np.cos(x[0])/(1+x[0]) - np.sin(x[0])/(1+x[0])**2 -((np.sin(x[1]))/(x[1] +1) - (np.sin(x[0])/(x[0] +1)))/(x[1]-x[0]))
  return y

def JF(x):
  y = np.zeros((2,2))  
  y[0,0] = ((-x[0]**2*np.sin(x[0])) -2*x[0]*np.sin(x[0]) - 2*x[0]*np.cos(x[0]) + np.sin(x[0]) - 2*np.cos(x[0]))/(x[0]+1)**3
  y[0,1] = (x[1]**2*np.sin(x[1]) + 2*x[1]*np.cos(x[1]) + 2*x[1]*np.sin(x[1]) + 2*np.cos(x[1]) - np.sin(x[1]))/((x[1] + 1)**3)

  y[1,0] = ((-np.sin(x[0])*(1+x[0]) - (np.cos(x[0])/((1+x[0])**2)) - np.cos(x[0])*(x[0]+1)-2*np.sin(x[0]))/((1+x[0])**3) - (-(np.cos(x[0])*(x[0]+1) - np.sin(x[0]))/((x[0]+1)**2)*(x[1] - x[0]) - (-1)*((np.sin(x[1])/(x[1]+1))) - np.sin(x[0])/(x[0]+1))/((x[1]-x[0])**2))
  y[1,1] = (((np.cos(x[1])*(x[1]+1)-np.sin(x[1]))/(x[1]+1)**2*(x[1]-x[0])) - ((np.sin(x[1]))/(x[1]+1) -(np.sin(x[0]))/(x[0]+1)))/(x[1]-x[0])**2
  return y
```

# Execução Das funções:

Chute inicial: ( 1 , 8 )

```python

x = np.array([1,8]) #CHUTE INICIAL

x=x-np.linalg.inv(JF(x)).dot(F(x)) #INTERAÇÃO 1
x=x-np.linalg.inv(JF(x)).dot(F(x)) #INTERAÇÃO 2
x=x-np.linalg.inv(JF(x)).dot(F(x)) #INTERAÇÃO 3
x=x-np.linalg.inv(JF(x)).dot(F(x)) #INTERAÇÃO 4
x=x-np.linalg.inv(JF(x)).dot(F(x)) #INTERAÇÃO 5
x=x-np.linalg.inv(JF(x)).dot(F(x)) #INTERAÇÃO 6

x #PRINT DO ENCONTRO ENTRA AS DUAS CURVAS
```
